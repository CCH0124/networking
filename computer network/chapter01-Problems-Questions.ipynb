第一條鏈路的速率為 $R_s$；第二條鏈路的速率為 $R_c$，且 $R_c < R_s$（第二鏈路是瓶頸）。第一個封包完成傳輸後，第二個封包會在第二鏈路等待，因為第二鏈路的速率較慢。

如果希望第二個封包不排隊（即到達第二鏈路時，第一個封包已完成在第二鏈路的傳輸），需要控制伺服器發送第二個封包的時間間隔 $T$。

- 第一個封包在第二鏈路的傳輸時間：$d_{trans\_scond} = \frac{L}{R_c}$
- 為了避免第二個封包排隊，第二個封包應在 $d_{trans\_second}   後到達第二鏈路，因此伺服器發送第二個封包的時間間隔 $T$ 必須滿足： $T >= d_{trans\_second} - d_{trans\_first}$，該 $d_{trans\_second} = \frac{L}{R_s}$

因此，最後須滿足 $T >= \frac{L}{R_c} - \frac{L}{R_s}$
